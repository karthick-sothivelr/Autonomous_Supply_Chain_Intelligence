```python
import requests
import json

# The ports we configured in 'run_vendors.sh'
VENDOR_PORTS = range(8001, 8012)  # 8001 to 8011

print(f"ğŸ” Scanning VeganFlow Vendor Ecosystem ({len(VENDOR_PORTS)} expected)...\n")

success_count = 0

for port in VENDOR_PORTS:
    url = f"http://localhost:{port}/.well-known/agent-card.json"
    try:
        response = requests.get(url, timeout=2)
        
        if response.status_code == 200:
            card = response.json()
            name = card.get('name', 'Unknown')
            description = card.get('description', 'No description')
            
            # Success output
            print(f"âœ… [Port {port}] {name}")
            print(f"   â””â”€ {description[:60]}...")
            success_count += 1
        else:
            print(f"âŒ [Port {port}] Error: HTTP {response.status_code}")
            
    except requests.exceptions.ConnectionError:
        print(f"âŒ [Port {port}] Connection Refused (Agent not running?)")
    except Exception as e:
        print(f"âš ï¸ [Port {port}] Unexpected Error: {e}")

print(f"\nğŸ“Š System Status: {success_count}/{len(VENDOR_PORTS)} Vendors Online")
```
